 ## les conditions  

les conditions en python retourne un boolean (True or False) selon le résultat de la comparaison.  

plus d'information : https://www.w3schools.com/python/python_conditions.asp

En python on défini une condition comme ceci : 

                                      a "comparateur" b
                              
les différents comparateurs:

    a < b : a plus petit que b
    a > b : a plus grand que b
    a == b : a est égal à b
    a >= b : a plus grand ou égal à b
    a <= b : a plus petit ou égal à b
    a != b : a pas égal à b
    
Les comparateurs est égal (==) et (!=) peuvent aussi être utilisés pour comparer 2 strings.

Il existe aussi des mots clefs important pour les conditions

- not : inverse la valeur de la comparaison 

- and : permet de regarder si les 2 comparaisons sont vraies voir "tableau vérité"

```python
a = 10
b = 20
x = 10

c = "test"
d = "test"
e = "hello"

print("1", a < b)
print(not ("2", a > b))
print("3", c == d)
print("4", c == e)
print("5", a==x)

print("6", c == d and x == b)
```

ces dernières servent à éxécuter certaines parties du programme selon la valeure d'une expression.

en python les conditions sont déclarées de cette sorte:

```python
if a > b:
    print("A est plus grand que b")
elif a==b:
    print("A est égal à b")
else:
    print("A est plus petit que B")
    
```

Le code à l'indentation 1 et en dessou de if est exécuter si et seulement si l'expression devant ce dernier est vrai, si elle est fausse il va contrôler la deuxième expression qui est déclarée avec elif, puis si cette dernière est vraie, si elle ne l'est pas la dernière condition exécutée va être else. 

Il est possible de créer des déclarations de condition très grande avec plusieurs elif. De type:

if (expression1):
    //code à exécuter si expression 1 vrai
elif (expression2):
    //second code à exécuter si expression 2 vrai
elif (expression3):
    //troisième code à exécuter si expression 3 vrai
    
    ...
else:
    //exécuter si aucune des conditions au dessus était vraie.

## les listes

une liste est un structure de donnée représentant une séquence d'élement cette dernière est modifiable et ordrée.

- une liste vide  s'instancie avec des parenthèses carrées [].

- cette dernière peut contenir de tout type de donnée, des strings, des intégrales, des autres listes(nested list)

- Une liste est ordrée c'est à dire que chaque élement est "stocké" à un index défini qui commence à 0 (premier élement de la liste)

```python
liste_1 = []

liste_2 = [1,2,3,4,5]

fruits = ["banane", "pomme", "fraise", "raisins"]

print(fruits[3])
print(fruits[0])
```

## les boucles for 

une boucle for est utilisé pour itérer dans une séquence par exemple une liste

pour chaque élement il est possible d'exécuter des actions sur chaque élement 1 à 1 sur les élements itérés.



```python
counter = 0
for fruit in fruits:
    print(fruit)
    counter += 1
```
