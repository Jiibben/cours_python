## Les déclarations de condition :

ces dernières servent à éxécuter certaines parties du programme selon la valeure d'une expression.

en python les conditions sont déclarées de cette sorte:

```python
if a > b:
    print("A est plus grand que b")
elif a==b:
    print("A est égal à b")
else:
    print("A est plus petit que B")
    
```

Le code à l'indentation 1 et en dessou de if est exécuter si et seulement si l'expression devant ce dernier est vrai, si elle est fausse il va contrôler la deuxième expression qui est déclarée avec elif, puis si cette dernière est vraie, si elle ne l'est pas la dernière condition exécutée va être else. 

### Conditions chaining

Il est possible de créer des déclarations de condition très grande avec plusieurs elif. De type:


```python
number = 5

if number > 3:
    print("le nombre est plus grand que 3")
elif number > 5:
    print("le nombre est plus grand")
elif number == 6:
    print("le nombre est égal à 6")
elif number > 6:
    print("le nombre est plus grand que 6")
else:
    print("le nombre est plus petit ou égal à 3")
```

## les boucles for 

une boucle for est utilisé pour itérer dans une séquence(itérable) par exemple une liste

pour chaque élement il est possible d'exécuter des actions sur chaque élement 1 à 1 sur les élements itérés.



```python
counter = 0
for fruit in fruits:
    print(fruit)
    counter += 1
```

il y'a plusieurs mot clefs qui existe pour intéragir avec les boucles notamment:
- continue: lorsque ce mots clefs est appelé un tour de boucle est sauté (pas d'action dans ce tour de boucle)
- break la boucle est cassé et s'arrête

### Exemple pour !continue! : 

```python
greets = ["salut", "bonjour", "hi", "coucou"]

for i in greets:
    if i == "hi":
        continue
    else:
        print(i)
```

on voit qu'au tour de boucle ou la variable i vaut hi la condition i=="hi" est évaluée à vrai donc ce qui est dans cette condition est exécuté soit //continue// et que le tour de boucle est sauté. (on imprime pas hi)

### Exemple pour !break!:

petite notes : 

range(0,10) crée une itérable avec toutes les valeurs entre 0 et 9 (pas 10 car non inclusif)

```python
debut = 0
print("début du code :)")

for i in range(0,10):
    if i > 5:
        break
    print(i)
    debut += 1
    
print("Fin du code")
```

on voit que dès que la variable debut atteint une valeur de plus de 5 la boucle est cassé et le code se termine.

## while loop 

la boucle while est une boucle qui "tourne" tant que la condition qu'on lui donne est vrai. 

#### Syntaxe en python:

```python
#while (condition)
    #<code à exécuter>
```

#### exemple : 

```python
nombre1 = 0

while nombre1 <6:
    print(nombre1)
    nombre1 = nombre1 + 1
```

On observe que tant que la condition est vrai (nombre1 <6) le code en dessou est exécuté soit imprimé nombre1 et l'incrémenter de 1.

/!\ attention /!\
Si la condition est toujours évaluée à vrai (True) la boucle est infinie.

# Les fonctions "classiques"

Il existe deux types de fonctions, les fonctions "classique" et les fonctions dites lambda. Une fonction est un bloc de code qui n'est exécuté uniquement lorsque la fonction en question est appelé.

plus d'info https://www.w3schools.com/python/python_functions.asp

#### syntaxe :

```python
#def nomdelafonction():
    #code de la fonction
```

```python
def say_hi():
    print("Hello world !")
    
say_hi()
say_hi()
say_hi()
```

la fonction est définie comme présenté plus haut. Pour appeler une fonction il faut utiliser son nom et ajouter les 2 parenthèse après comme ceci : nomdelafonction()

#### fonctions à arguments

une fonction peut prendre un ou plusieurs arguments, ces derniers peuvent être par la suite utiliser dans la fonction.

##### exemple 1:

```python
def say_word(word):
    print(word)
    
say_word("bonjour")
say_word("salut")
```

##### exemple 2:

```python
def puissancecarre(n1):
    print(n1 ** 2)
    
puissancecarre(10)
puissancecarre(3)
```

#### valeurs de retour d'une fonction

si on essaie de print() une fonction qui ne retourne aucune valeur on n'obtiendra un ensemble vide (None en python)

#### exemple:

```python
def coucou():
    print("Coucou, comment ça va ?")
    
print(coucou())
```

On voit que la fonction est exécutée mais il n'y a pas de valeur de retour donc quand on imprime la fonction on obtient la valeur None.

pour ajouter une valeur de retour on utilise le mot clefs return

#### exemple:

```python
def coucou():
    return "coucou comment ça va ?"

coucou()
coucou()

print(coucou())
```

on voit que même si on exécute coucou 2 x sans print() son résultat on obtient rien ceci est dû au fait que coucou retourne uniquement "coucou comment ça va ? " donc il faut imprimer sa valeut de retour comme à la fin.

Une utilisation fréquente et utile des fonctions est d'assigner sa valeur de retour à une variable

#### exemple:

```python
def puissancecube(n1):
    return n1 ** 3

valeur1 = puissancecube(2)

valeur2 = puissancecube(4)

print(valeur1)
print(valeur2)

print(puissancecube(6))
```

## les dictionnaires
