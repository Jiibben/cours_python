- les variables sont "case sensitive" c'est à dire que AGE et age ne sont pas les mêmes

## opérateurs mathématiques

python supporte toutes les opérations mathématique de base:
   - addition : +
   - soustraction : -
   - division : /
   - multiplication : *
   - puissance : **
   - modulo : % (retourne le reste d'une division)
   
plus d'information : https://www.w3schools.com/python/python_operators.asp

```python
print("addition:", 5+3)
print("soustraction:", 5-3)
print("division:", 6/3)
print("multiplication:", 5*3)
print("puissance:", 5**3)
print("modulo:", 15 % 4)
```

L'opérateur + et * sont aussi utilisable sur des chaînes des caractères :

```python
print("aa" * 6)
print("ab" + "cd")
```

 ## les conditions  

les conditions en python retourne un boolean(True or False) selon l'expression 

plus d'information : https://www.w3schools.com/python/python_conditions.asp

En python on défini une condition comme ceci : 

                              a expression b
                              
les différentes :

    a < b : a plus petit que b
    a > b : a plus grand que b
    a == b : a est équivalent à b
    a >= b : a plus grand ou égal à b
    a <= b : a plus petit ou égal à b
    a != b : a pas égal à b



```python
a = 10
b = 20
x = 10

c = "test"
d = "test"
e = "hello"

print("1", a < b)
print("2", a > b)
print("3", c == d)
print("4", c == e)
print("5", a==x)
```
